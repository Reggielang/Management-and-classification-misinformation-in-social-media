# The processing of  Chinese tweet text data includes：
1.Remove handles and special symbols

2.Import Chinese stop words and delete them

3.Import Chinese word segmentation toolkit and make word segmentation for tweets

```python
tweet_chinese = pd.read_csv('tweet_chinese.csv')
```

```python
# Filter out some useless handles
import re
from string import digits
from zhon.hanzi import punctuation
reg_map = {
# #     re.compile("rt [@0-9a-z_]{0,10}:"),
#     re.compile("[a-zA-z]+://[^\s]*"),
#     re.compile('http|htttps…'),
#     re.compile('[a-z]*[:.]+\S+'),
#     re.compile("@[0-9a-z]+"),
# #     re.compile("rt|rt :[\s]+"),
#     re.compile("#"),
#     re.compile('https?:\/\/\S+'),
#     re.compile('[\f\n\r\t\v]'),
# #     re.compile(r"\[\S+\]"),
#     re.compile('[^a-zA-Z.,;《》？！“”‘’@#￥%…&×（）——+【】{};；●，。&～、|\s:：'+'\u4e00-\u9fa5]+'),
#     re.compile('[；]+'),
#     re.compile('[a-zA-Z]'),
#     re.compile(':|;'),
#     re.compile(r"[%s]+" %punctuation),
#     re.compile(r"[%s]+" %digits),
    re.compile(' '),
    re.compile('[^\u4E00-\u9FD5]+'),
    
}
def lower_and_remove_with_reg(text: str) -> str:
#     text = text.lower()
    for v in reg_map:
        text = v.sub("", text)
    return text
```

```python
# Filter out some useless handles
tweet_chinese['text']=tweet_chinese['tweet'].apply(lower_and_remove_with_reg)
tweet_chinese['text'].head()
```

```python
tweet_chinese['text']=tweet_chinese['text'][tweet_chinese['text'].str.len() > 1]
tweet_chinese['text'].head()
```

```python
tweet_chinese.head()
```

```python
tweet_chinese.dropna(subset=['text'],inplace=True)
```

```python
#导入中文停用词库，并删除停用词
stopwords = [line.rstrip() for line in open('./baidu_stopwords.txt', 'r', encoding='utf-8')]
tweet_chinese['word'] = tweet_chinese['text'].apply(lambda sen:" ".join(x for x in sen.split() if x not in stopwords))
tweet_chinese['word'].head()
```

```python
#导入中文分词工具包，并对tweet进行分词
import jieba
tweet_chinese['word'] = tweet_chinese['word'].apply(lambda x: ' '.join(jieba.cut(x)))
tweet_chinese['word'].head()
```

```python
tweet_chinese_mining = tweet_chinese[['tweet_id','text','word','user_location','tweet']]
```

```python
tweet_chinese_mining.text.isna().value_counts()
```

```python
tweet_chinese_mining.dropna(subset=['text'],inplace=True)
```

```python
#全部字体转化为简体中文
import zhconv
tweet_chinese_mining.text = tweet_chinese_mining.text.apply(lambda x : zhconv.convert(x,'zh-cn'))
```

```python
tweet_chinese_mining.head()
```

# Use multiple sentiment dictionary matching toolkits to match negative Chinese Twitter
1.中文情感分析库(Chinese Sentiment))可对文本进行情绪分析、正负情感分析 reference : https://github.com/thunderhit/cnsenti
 1. 情感分析默认使用的知网Hownet


2.bixin
 2. Chinese Sentiment Analysis base on dictionary and rules. 

```python
from cnsenti import Sentiment
senti = Sentiment()
```

```python
tweet_chinese_mining['sentiment'] = tweet_chinese_mining['text'].apply(senti.sentiment_count)
tweet_chinese_mining.head()
```

```python
#The dictionary expands to the column of dataframe
df_sentiment = tweet_chinese_mining['sentiment'].apply(pd.Series)
```

```python
tweet_chinese_mining['sentiment'] = df_sentiment.pos - df_sentiment.neg
```

```python
tweet_chinese_mining.head()
```

# bixin package
reference : https://github.com/bung87/bixin
1. sentiment dictionary matching toolkits

2. Emotion range is a number from -1 to 1

```python
from bixin import predict
```

```python
tweet_chinese_mining['sentiment_v2'] = tweet_chinese_mining.text.apply(predict)
```

```python
tweet_chinese_mining
```

```python
#留下消极情感的推特文本
tweet_chinese_negative = tweet_chinese_mining[(tweet_chinese_mining.sentiment < 0) | (tweet_chinese_mining.sentiment_v2 < 0)]
```
